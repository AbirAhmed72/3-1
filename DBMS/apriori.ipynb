```python
import pandas as pd

df = pd.DataFrame(pd.read_csv('D:\\Poralekha\\3-1\\DBMS\\apriori.csv', delimiter=':'))


# for i in df['TID']:
#     print(i.split(','))

data = []

# Iterate over the rows of the DataFrame
for _, row in df.iterrows():
    # Extract the TID and PRODUCTS values
    tid = row['TID']
    products = row['PRODUCTS'].split(',')

    # Store the values in a list and append it to the new list
    data.append([tid, products])

# print(data)

transactions_dict = {}
for entry in data:
    tid = entry[0]
    products = entry[1]
    transactions_dict[tid] = products

print(transactions_dict['TID1'][1])

```
