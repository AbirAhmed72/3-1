```python
import pandas as pd
import numpy as np
import random
from sklearn.utils import shuffle
# from google.colab import drive
import statistics
import math
from scipy.stats import norm
# drive.mount('/content/drive')

accuracy = 0

for i in range(10):
    true_positive = 0
    true_negative = 0
    false_positive = 0
    false_negative = 0

    dataset = pd.DataFrame(pd.read_excel('Cryotherapy.xlsx'))
    dataset = shuffle(dataset)
#   print(dataset)

    total_yes = 0
    total_no = 0
    age_mean = 0
    age_tot = 0
    age_dev = 0
    time_mean = 0
    time_tot = 0
    time_dev = 0
    wart_mean = 0
    wart_tot = 0
    wart_dev = 0
    area_mean = 0
    area_tot = 0
    area_dev = 0
    males = 0
    females = 0
    male_yes = 0
    male_no = 0
    female_yes = 0
    female_no = 0
    type_1 = 0
    type_2 = 0
    type_3 = 0
    type_1_yes = 0
    type_1_no = 0
    type_2_yes = 0
    type_2_no = 0
    type_3_yes = 0
    type_3_no = 0

    patient = dataset.values.tolist()
    # type: ignore
    # print(patient)
    # print(patient[0][1])

    train = []

    for i in range(72):
        col = []
        for j in range(7):
            col.append(patient[i][j])
        train.append(col)

    test = []
    for i in range(72, 90):
        col = []
        for j in range(7):
            col.append(patient[i][j])
        test.append(col)

    patient_age = []
    patient_time = []
    patient_wart = []
    patient_area = []

    for i in range(72):
        if train[i][0] == 1:
            males += 1
            if train[i][6] == 1:
                male_yes += 1
            else:
                male_no += 1
        else:
            females += 1
            if train[i][6] == 1:
                female_yes += 1
            else:
                female_no += 1

    for i in range(72):
        if train[i][4] == 1:
            type_1 += 1
            if train[i][6] == 1:
                type_1_yes += 1
            else:
                type_1_no += 1
        elif train[i][4] == 2:
            type_2 += 1
            if train[i][6] == 1:
                type_2_yes += 1
            else:
                type_2_no += 1
        else:
            type_3 += 1
            if train[i][6] == 1:
                type_3_yes += 1
            else:
                type_3_no += 1

  # print(type_1,type_2,type_3,type_1_yes,type_1_no,type_2_yes,type_2_no,type_3_yes,type_3_no)

  # print(male,female,male_no,male_yes,female_yes,female_no)

    for i in range(72):
        patient_age.append(patient[i][1])
    for i in range(72):
        patient_time.append(patient[i][2])
    for i in range(72):
        patient_wart.append(patient[i][3])
    for i in range(72):
        patient_area.append(patient[i][5])

    age_dev = statistics.stdev(patient_age)
    time_dev = statistics.stdev(patient_time)
    wart_dev = statistics.stdev(patient_wart)
    area_dev = statistics.stdev(patient_area)

    for i in range(72):
        age_tot += patient[i][1]
        age_mean = age_tot/72

    for i in range(72):
        time_tot += patient[i][2]
        time_mean = time_tot/72

    for i in range(72):
        wart_tot += patient[i][3]
        wart_mean = wart_tot/72

    for i in range(72):
        area_tot += patient[i][5]
        area_mean = area_tot/72

    for i in range(72):
        if patient[i][6] == 1:
            total_yes += 1
        else:
            total_no += 1

        prob_yes = total_yes/72
        prob_no = total_no/72

  # testing

    for i in range(72, 90):
        prob_tot_yes = 1
        prob_tot = 1

        prob_age = norm.pdf(patient[i][1], age_mean, age_dev)
        prob_time = norm.pdf(patient[i][2], age_mean, age_dev)
        prob_wart = norm.pdf(patient[i][3], age_mean, age_dev)
        prob_area = norm.pdf(patient[i][5], age_mean, age_dev)
        # print(prob_age,age_mean,age_dev,patient[i][1])

        prob_tot_yes *= ((prob_time*prob_age*prob_wart *
                         prob_area))*(total_yes/72)
        prob_tot *= prob_area*prob_age*prob_time*prob_wart
        if patient[i][0] == 1:
            prob_tot_yes *= male_yes/total_yes
            prob_tot *= males/72
            if patient[i][4] == 1:
                prob_tot_yes *= type_1_yes/total_yes
                prob_tot *= type_1/72
            elif patient[i][1] == 2:
                prob_tot_yes *= type_2_yes/total_yes
                prob_tot *= type_2/72
            else:
                prob_tot_yes *= type_3_yes/total_yes
                prob_tot *= type_3/72

        else:
            prob_tot_yes *= female_yes/total_yes
            prob_tot *= females/72
            if patient[i][4] == 1:
                prob_tot_yes *= type_1_yes/total_yes
                prob_tot *= type_1/72
            elif patient[i][1] == 2:
                prob_tot_yes *= type_2_yes/total_yes
                prob_tot *= type_2/72
            else:
                prob_tot_yes *= type_3_yes/total_yes
                prob_tot *= type_3/72

        prob_final = prob_tot_yes/prob_tot
        # print(prob_final)
        if prob_final > 0.5:
            if patient[i][6] == 1:
                true_positive += 1
            else:
                false_positive += 1
        else:
            if patient[i][6] == 1:
                false_negative += 1
            else:
                true_negative += 1

    accuracy += ((true_positive+true_negative)/(true_positive +
                 false_positive+true_negative+false_negative))
    #   print(accuracy)
    
print(f"Accuracy % is {accuracy*10}")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import confusion_matrix

dataframe = pd.read_excel('/content/drive/MyDrive/5thSemester/DBMS_2/Cryotherapy.xlsx')
# dataframe.head()
# dataframe.describe()
# dataframe['Result_of_Treatment'].value_counts()
# dataframe.groupby('Result_of_Treatment').mean()
x = dataframe.drop('Result_of_Treatment', axis=1)
# print(x)
y = dataframe['Result_of_Treatment']
# print(y)
scaler = StandardScaler()
scaler.fit(x)
std_data = scaler.transform(x)
x = std_data
# print(x)
y = dataframe['Result_of_Treatment']
# print(y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = True)
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
print(y_pred)
print(metrics.accuracy_score(y_test, y_pred))
cm = np.array(confusion_matrix(y_test,y_pred))
print(cm)
```
